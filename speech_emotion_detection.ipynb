```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from glob import glob
import librosa
import librosa.display
import IPython.display as ipd
from itertools import cycle
```

```python
audio=glob('archive/audio_speech_actors_01-24/Actor_*/*.wav')
ipd.Audio(audio[0])
```

```python
y, sr = librosa.load(audio[0])
print("y: ", y[:10])
print(y.shape)
print(sr)
```

```python
a1 = pd.Series(y)
a1.plot(figsize=(10, 5),
       lw=1,
       title='Raw Audio')
plt.show()
```

```python
y_trimmed, _ = librosa.effects.trim(y, top_db=20)
pd.Series(y_trimmed).plot(figsize=(10,5),
                         lw=1,
                         title="Raw audio trimmed")
plt.show()
```

```python
Y = librosa.stft(y)
Ydb = librosa.amplitude_to_db(abs(Y))
plt.figure(figsize=(20, 5))
librosa.display.specshow(Ydb, sr=sr, x_axis='time', y_axis='hz')
plt.title('Spectrogram')
plt.colorbar()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load the audio file
y, sr = librosa.load(audio[0])
plt.figure(figsize=(20, 4))
# Preprocessing step 1: Mel-frequency Cepstral Coefficients (MFCCs)
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('MFCCs')
plt.show()
plt.figure(figsize=(20, 4))
# Preprocessing step 2: Mel-scaled spectrogram
mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)
librosa.display.specshow(librosa.power_to_db(mel_spec, ref=np.max), y_axis='mel', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-scaled Spectrogram')
plt.show()
plt.figure(figsize=(20, 4))
# Preprocessing step 3: Chromagram
chroma = librosa.feature.chroma_stft(y=y, sr=sr)
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plt.colorbar()
plt.title('Chromagram')
plt.show()
plt.figure(figsize=(20, 4))
# Preprocessing step 4: Spectral contrast feature
spectral_contrast = librosa.feature.spectral_contrast(y=y, sr=sr)
librosa.display.specshow(spectral_contrast, x_axis='time')
plt.colorbar()
plt.title('Spectral Contrast Feature')
plt.show()
plt.figure(figsize=(20, 4))
# Preprocessing step 5: Tonnetz representation
tonnetz = librosa.feature.tonnetz(y=librosa.effects.harmonic(y), sr=sr)
librosa.display.specshow(tonnetz, y_axis='tonnetz', x_axis='time')
plt.colorbar()
plt.title('Tonnetz Representation')
plt.show()
```
